```python
import pandas as pd

df = pd.read_csv("../data/raw_products.csv")
```

```python
df
```

```python
columns_to_drop = ['images', 'thumbnail', 'meta'] 
df = df.drop(columns=columns_to_drop)
```

```python
# If tags are strings like "tag1, tag2", convert them to lists
df['tags'] = df['tags'].apply(lambda x: x if isinstance(x, list) else x.split(', '))
```

```python
# Create a new column 'subcategory' containing only the second tag
df['subcategory'] = df['tags'].apply(lambda x: x[1] if len(x) > 1 else None)
```

```python
print(df[['id','category','subcategory']])
```

```python
# Remove leading/trailing quotes and spaces
df['subcategory'] = df['subcategory'].astype(str).str.strip().str.replace('"', '').str.replace("'", "").str.replace(']', '').str.replace('[', '').str.replace('womens watches', 'watches')
```

```python
print(df[['id','category','subcategory']])
```

```python
df
```

```python
df = df.drop(columns=['tags'])
```

```python
df
```

```python
import ast  # safely evaluate string to Python object

# Convert strings to dictionaries
df['dimensions'] = df['dimensions'].apply(ast.literal_eval)
```

```python
dimensions = pd.json_normalize(df['dimensions'])
dimensions['product_id'] = df['id']

# Optional: reorder columns
dimensions = dimensions[['product_id'] + [col for col in dimensions.columns if col != 'product_id']]

print(dimensions.head())
```

```python
df
```

```python
df['width'] = dimensions['width']
df['height'] = dimensions['height']
df['depth'] = dimensions['depth']

```

```python
df
```

```python

print(df['reviews'].iloc[0])
```

```python
def get_first_review(review_list):
    if isinstance(review_list, list) and len(review_list) > 0:
        review = review_list[0]  # take the first dict in the list
        return pd.Series({
            'rating': review.get('rating'),
            'comment': review.get('comment')
        })
    else:
        return pd.Series({'rating': None, 'comment': None})

df[['rating','comment']] = df['reviews'].apply(get_first_review)
```

```python
df = df.drop(columns=['reviews', 'reviews_dict'])

```
